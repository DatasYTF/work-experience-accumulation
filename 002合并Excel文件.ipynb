```python
import pandas as pd
import os
import numpy as np
import time

start_time = time.time()

# 文件夹所在路径
路径 = "D:/001例行工作/每日例行/【ERP统计/【PBI\物流ERP基础数据/001订单明细（0点-24点）/"
# 新建空白DataFrame
合并表 = pd.DataFrame()

# os.listdir()将文件夹内文件名变为list
for 文件名 in os.listdir(路径):
    表格 = pd.read_excel(路径 + 文件名,dtype={'运单号':object,'受理时间':np.datetime64,
                                        "发件人手机":object,'发件人电话':object,
                                        "客户批次号":object,"收件人电话":object,
                                       "收件人手机":object,"数量":object})
    合并表 = pd.concat([合并表,表格])

end_time = time.time()
print("合并用时{}s".format(end_time - start_time))
```

```python
df = pd.read_excel(路径+"]半天.xlsx",dtype={'运单号':object,'受理时间':np.datetime64,
                                        "发件人手机":object,'发件人电话':object,
                                        "客户批次号":object,"收件人电话":object,
                                       "收件人手机":object,"数量":object})
a = df.columns.to_list()
data = 合并表[a]
datas = data.dropna(subset = ["受理时间"])
```

```python
datas = datas.sort_values(by = '受理时间',ascending=False) # 按受理时间 倒序
datas = datas.drop_duplicates(subset=["运单号"]) # 运单号去重——保留第一次出现的。
```

```python
start_time = time.time()
datas.to_excel('./文件/202107月订单明细.xlsx',index=False,encoding='gbk')
# datas.to_csv('./文件/001订单明细.csv',index=False)

end_time = time.time()
print("保存用时{}s".format(end_time - start_time))
```
